```python
import pandas as pd

import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# Curve Plot

x = np.linspace(0, 10, 50)
y = np.sin(x)

plt.plot(x, y)
plt.title("Sin wave")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
# Line Plot
x = np.linspace(0, 10, 50)
y = 4 * x
plt.plot(x, y)
plt.title("Line Plot")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
# Line plot with some more arguments
x = np.linspace(1, 10, 10)
y = 4 * x
plt.plot(x, y, marker='o', linestyle='--', color='r', ms=5) # ms: marker size
plt.title("Line Plot")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
# Bar Plot

products = ['Iphone 17', 'Samsung Galaxy 530', 'Google Pixel 10', 'OnePlus 12', 'Sony Xperia 1 V']
sales = [150, 120, 90, 80, 70]

plt.figure(figsize=(15, 6))
plt.bar(products, sales, color='skyblue')

plt.title('Smart phone sales 2025')
#plt.xticks(rotation=90)
plt.xlabel('Products')
plt.ylabel('Unit solve in thousand')
plt.show()
```

```python
# Pie Plot

products = ['Iphone 17', 'Samsung Galaxy 530', 'Google Pixel 10', 'OnePlus 12', 'Sony Xperia 1 V']
sales = [150, 120, 90, 80, 70]

plt.pie(x=sales, labels=products, autopct='%1.1f%%')
plt.title('Smartphone Sales Distribution 2025')
plt.show()
```

```python
# Pie Plot with some more argument

products = ['Iphone 17', 'Samsung Galaxy 530', 'Google Pixel 10', 'OnePlus 12', 'Sony Xperia 1 V']
sales = [150, 120, 90, 80, 70]

plt.pie(x=sales, labels=products, autopct='%1.1f%%', explode=[0.1, 0.1, 0, 0, 0])
plt.title('Smartphone Sales Distribution 2025')
plt.show()
```

```python
#Scatter plot (biểu đồ phân tán/phân phối)
reading_scores = np.random.normal(75, 10, 200) #(mean: trung bình, std: độ lêch chuẩn, size: tạo 200 giá trị)
writing_scores = reading_scores + np.random.normal(0, 5, 200)#(mean: trung bình, std: độ lêch chuẩn, size: tạo 200 giá trị)
plt.scatter(reading_scores, writing_scores)
plt.title("IL Score")
plt.xlabel("Reading score")
plt.ylabel("Writting score")
plt.show()
```

```python
# Histogram mathplotlib
x = np.random.normal(0, 10, 100)
plt.hist(x, bins=15, color='skyblue', edgecolor='black')
plt.title('Histogram of normal distributted data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# Histogram seaborn
x = np.random.normal(0, 10, 100)
sns.histplot(x, bins=15, color='skyblue', kde=True)
plt.title('Histogram of normal distributted data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
