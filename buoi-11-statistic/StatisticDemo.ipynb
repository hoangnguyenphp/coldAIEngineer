Kết quả - outcome: xúc sắc 1, 2, 3, 4, 5, 6
Biến cố - event: tập hợp nhiều kết quả, xúc sắc chắn {2, 4, 6}, xúc sắc lẻ {1, 3, 5}

Xác suất của 1 biến cố (event)
    Luôn nằm từ 0 - 1 
    Xác suất có thể ở dạng %
        0 -> không bao giờ xảy ra;
            vd: không thể có biến cố tung đồng xu vừa được mặt xấp vừa được mặt ngửa
        1 -> chắc chắn xảy ra
            vd: biến cố tung đồng xu sẽ được mặt xấp hoặc ngửa
        
        Ỏ giữa 0 và 1
            Nếu xác suất gần 0 (ví dụ: 0.05 hoặc 5%) -> Biến cố rất ít khả năng xảy ra.
            
            Nếu xác suất gần 1 (ví dụ: 0.95 hoặc 95%) -> Biến cố có khả năng cao xảy ra.

            Nếu xác suất = 0.5 (50%) -> Biến cố có cơ hội xảy ra ngang bằng với không xảy ra.

    Ứng dụng: dựa trên khả năng xảy ra để đưa ra quyết định tốt hơn.

Quy tắc cơ bản trong xác suất
    Công thức P(A) = Số kết quả mong muốn / Tổng số kết quả có thể xảy ra
        P: Probability Notation
        (A), (B): đại diện cho biến cố

    Ví dụ:
        Tung đồng xu
            P(mặt ngửa) = 1 / 2 = 0.5

        Tung xúc sắc
            P(nhỏ hơn 3) = 2/6 = 0.333 = 33.33 %

        * Xác suất dài hạn: không có nghĩa là 10 lần tung đồng xu là sẽ 5 lần được mặt ngửa có thể là 6 ngửa 4 sấp hoặc 7 sấp 3 ngửa. Nhưng nếu tung nhiều lần hơn thì sẽ tiến về 0.5 (50%)

    Ba loại biến cố trong xác suất
        Quy tắc bổ sung (Complement Rule) và Quy tắc cộng (Additional Rule) áp dụng cho các biến cố loại trừ nhau

            - Biến cố loại trừ nhau: không thể tung xúc sắc vừa ra biến cố các số chẵn vừa ra biến cố các số lẻ

            - Quy tắc bổ sung (Complement Rule): tổng xác suất và xác suất biến cố bổ sung của nó (phần bù) luôn bằng 1 (lấy tổng khả nay xảy ra của các biến cố trừ đi % biến cố đã biết để tính được % của biến cố còn lại)
                P(A') = 1 - P(A) => P(A') + P(A) = 1

                Vd: Xác suất ngày mai có tuyết là 40 %
                    P(không có tuyết) = 1 - P(có tuyết) = 1 - 0.4 = 0.6

            - Quy tắc cộng (Additional Rule): 
                Ví dụ: tính xác suất tung xúc xắc để được mặt 2 và 4
                    P(2 hoặc 4) = 1/6 + 1/6 = 1/3

        Quy tắc nhân (Mutiplication Rule) áp dụng cho các biến cố độc lập (thực hiện thí nghiệm nhiều lần hoặc nhiều thí nghiệm khác nhau và mỗi thí nghiệm thực hiện 1 lần)
            Biến cố độc lập: 
                biến cố tung xúc sắc được mặt lẻ ở lần tung trước không ảnh hưởng đến kết quả của lần tung tiếp theo => thực hiện thí nghiệm nhiều lần

                ví dụ tính xác suất 1 người nghỉ làm vào ngày mưa, thì xác biến cố nghỉ làm và biến cố trời mưa độc lập với nhau => nhiều thí nghiệm khác nhau

            P(A và B) = P(A) x P(B)
                Tính biến cố tung xúc xắc 2 lần liên tiếp, 1 lần được mặt 1, một lần được mặt 6
                    P(1 và 6) = 1/6 x 1/6 = 1/32



Định lý Bayes và Sự Kiện Phức Tạp
    Xác suất có điều kiện(Conditional Probbability): áp dụng cho các biến cố phụ thuộc nhau
        Hai biến cố là phụ thuộc nhau nếu kết quả của biến cố đầu tiên ảnh hưởng đến kết quả của biến cố thứ 2

        P(A và B) = P(A) x P(B|A) => Xác suất xảy ra B khi A xảy ra
        => P(B|A) = P(A và B) / P(A)
            |: ký hiệu biểu hiện sự phụ thuộc 

        Ví dụ 1: không tráo bài tính xác suất bốc bài 2 lần liên tiếp đều là con cơ
            P(lần 1 cơ và lần 2 cơ) = 13/52 x 12/51 = 1/17

        Ví dụ 2 Tình huống:
            Một cửa hàng bán lẻ trực tuyến có thống kê sau:
            · 20% khách hàng chi tiêu từ $100 trở lên.
            · Nếu khách hàng chi tiêu từ $100 trở lên, họ đủ điều kiện để nhận phiếu quả tặng miễn phí.
            · 10% khách hàng chi tiêu từ $100 trở lên nhận được phiếu quà tặng

        + Câu hỏi:
            Hỏi xác suất để một khách hàng vừa chỉ tiêu từ $100 trở lên vừa nhận được phiếu quả tặng?

                => P(A và B) = P(A) x P(B|A) = 0.2 x 0.1 = 0.02 = 2%


    Định lý Bayes cung cấp một cách để cập nhật xác suất của một biến cố dựa trên thông tin mới về biến cố đó.

        Xác suất tiên nghiệm (Prior): xác suất của một biến cố trước khi có dữ liệu mới
        Xác suất hậu nghiệm (Posterior): xác suất của một biến có sau khi có dữ liệu mới

        => Định lý Bayes giúp tính xác suất hậu nghiệm bằng cách cập nhật xác suất tiên nghiệm dựa trên dữ liệu quan sát được

        P(A|B) = (P(B|A) x P(A))/P(B)

            + Ý nghĩa của từng thành phần trong công thức:
                . P(A): Xác suất tiên nghiệm (Prior Probability) - Xác suất xảy ra biến có A trước khi có thông tin về B.
                · P(A|B): Xác suất hậu nghiệm (Posterior Probability) -> Xác suất xảy ra A khi biết B đã xảy ra.
                · P(B|A): Khả năng (Likelihood) -- Xác suất B xảy ra khi A đã xày ra.
                · P(B): Bằng chứng (Evidence) -· Xác suất xảy ra B trong mọi trường hợp.


         Ví dụ: Bộ lọc thư rác (Spam Filter) và Định lý Bayes
            Một trong những ứng dụng phố biến của định lý Bayes trong thế giới số là bộ lọc thư rác (spam filter).
            giúp dự đoán một email có phải là spam hay không.
            + Cách bộ lọc thư rác hoạt động:
            · Một bộ lọc spam nắng cao sẽ xem xét nhiều yếu tố, bao gồm:
            Nội dung email
            Tiêu de email
            Email có tệp đinh kèm không
            Loại miên của địa chỉ người gửi (.edu, .org. .com)

            + Bải toán đơn giản hóa:
            · Ta chỉ xem xét một yếu tố: sự xuất hiện của từ "money" trong email.

            + Bước 1: Xác định thông tin đã có
                Giá sử bạn thu thập được các dữ liệu sau:
                · Xác suất một email là spam: P(Spam) = 0.2 (20%)
                · Xác suất từ "money" xuất hiện trong một email bất kỳ: P(Money) = 0.15 (15%)
                · Xác suất từ "money" xuất hiện trong email spam: P(Money|Spam) = 0.4 (40%)


            + Bước 2: Xác định biến co A và B
                . Biến co A: Email là spam -- P(Spam)
                . Biến co B: Email chứa từ "money" -- P(Money)

                + Công thức Định lý Bayes:

                    P(A|B) =  P(B|A) x P(A) / P(B)

                + Thay vào công thức với biến cố spam và từ "money":
                    P(Spam | Money)= P(Money | Spam)x P(Spam) / P(Money) = 0.4 * 0.2 / 0.15 = 0.53 ~ 53%

        Ví dụ: + Tình huống thực tế
            Một bệnh viện đang sử dụng một xét nghiệm để chấn đoán bệnh X. Dữ liệu thống kẻ về xét nghiệm
            này cho thấy:
            · Xác suất một người bất kỳ trong dân số mặc bệnh X: P(Bệnh) = 0.1 (10%)
            · Xác suất một người có bệnh X nhận được kết quả xét nghiệm dương tính (độ nhạy của xét
            nghiệm): P(Dương tính | Bệnh) = 0.95 (95%)
            · Xác suất một người không mắc bệnh X nhưng vẫn có kết quả dương tính (tức là xét nghiệm cho
            kết quả sai, hay dương tính giá): P(Dương tính Không bệnh) = 1 - P(Dương tính Bệnh) = 0.05 (5%)

            + Câu hỏi: Nếu một người có kết quả xét nghiệm dương tính, xác suất thực sự mắc bệnh X của họ là
            bao nhiêu?



                + Bước 1: Xác định các biến cố
                    · Biến cố A: Người đó mắc bệnh X -- P(Bệnh) = 0.01
                    · Biến cố B: Kết quả xét nghiệm dương tính - P(Dương tính)
                    · Xác suất căn tìm: P(Bệnh | Dương tính) => tức là xác suất thực sự mắc bệnh khi có kết quả xét nghiệm dương tính.

                + Áp dụng Định lý Bayes:

                    P(A|B) = P(B|A) × P(A) / P(B)

                    P(Bệnh | Dương tính) = P(Dương tính | Bệnh) x P(Bệnh) / P(Dương tính)


                + Bước 2: Tính xác suất P(Dương tính) (Xác suất tổng của một kết quả
                dương tính)
                    Xác suất một người bất kỳ nhận được kết quả dương tính có thể xảy ra trong hai trường hợp:
                        Họ mắc bệnh X và xét nghiệm cho kết quả đúng (dương tính thật).
                        Họ không mắc bệnh nhưng xét nghiệm cho kết quả sai (dương tính giả).
                    Tổng xác suất của biến cố "dương tính" là:

                        P(Dương tính) = P(Dương tính | Bệnh) x P(Bệnh) + P(Dương tính | Không bệnh) x P(Không bệnh) = (0.95×0.01)+(0.05×0.99) = 0.0095+0.0495 = 0.059

                    + Xác suất nhận được kết quả xét nghiệm dương tính là 5.9%.

                
                Bước 3: Tính xác suất hậu nghiệm 
                    P(Bệnh | Dương tính) =  0.95 × 0.01 / 0.059 =  0.0095 /  0.059 = 0.161 = 16.1%

                    + Kết quả cuối cùng: Xác suất thực sự mắc bệnh khi có kết quả xét nghiệm dương tính là 16.1%.

        Sự khác biệt giữa P(Dương tính | Bệnh) và P(Bệnh | Dương tính)
            Trong bài toán xét nghiệm bệnh, có hai xác suất dễ bị nhầm lẫn:
            1. Xác suất một người xét nghiệm dương tính khi họ thực sự mắc bệnh:

                                P(Dương tính | Bệnh)

                Đây là độ nhạy của xét nghiệm, tức là nếu một người thực sự có bệnh, xét nghiệm sẽ phát hiện ra bệnh với xác suất bao nhiêu.

            2. Xác suất một người thực sự mắc bệnh khi nhận kết quả xét nghiệm dương tính:

                                 P(Bệnh | Dương tính)

                Đây là xác suất hậu nghiệm (posterior probability), tức là nếu một người đã nhận được kết quả dương tính, khả năng thực sự họ mắc bệnh là bao nhiêu.
            
            Hai xác suất này hoàn toàn khác nhau!

Phân Phối Xác Suất Rời Rạc và Ứng Dụng Trong Khoa Học Dự Liệu
    Các chuyên gia dữ liệu (data professionals) sử dụng phân phối xác suất để mô hình hóa dữ liệu và tìm ra các mô hình quan trọng.
    + Phân phối xác suất là gì?
        . Mô tà xác suất xảy ra của các kết quả khác nhau của một biến cố ngẫu nhiên.
        · Phân phối xác suất rời rạc áp dụng cho biến ngẫu nhiên rời rạc (discrete random variables).

    + Đặc điểm của biến ngẫu nhiên rời rạc:
        · Chỉ có thể nhận các giá trị nguyên.
        · Ví dụ: Kết quả của một lần tung xúc xắc có thể là 1, 2, 3, 4, 5 hoac 6 - không thể có số thập phân như 2.575 hoặc 3.184.
                    

    4 phân phối xác suất rời rạc quan trọng:
        Phân phối đồng đều (Uniform Distribution)
            -tung xúc xắc 1 lần -> khả năng xảy ra mặt 1, 2, 3, 4, 5, 6 đều bằng nhau là 1/6  

        Phân phối nhị thức (Binomial Distribution)
            - Áp dụng cho các sự kiện chỉ có thể có 2 biến cố và loại trừ nhau (mỗi lần thử chỉ có thể ra 1 biến cố) + thử nhiều lần + lần thử trước không ảnh hưởng đến lần thử sau

            - Ví dụ: Tung đồng xu 10 lần -> biểu đồ thể hiện xác suất ra mặt ngửa trong 10 lần tung: 0 có mặt ngửa nào, 1 lần ngửa, 2 lần ngửa...

            - Ứng dụng: xác suất 1 loại thuốc gây tác dụng phụ trên n người khác nhau; xác suất 1 giao dịch gian lận trên n giao dịch; xác suất cổ phiều tăng giá; phân loại ảnh trong machine learning (một ảnh có phải là mèo không)...

            - Công thức P(X=k) = (n!/k! x (n-k)!​) x  p^k x (1−p)^n−k
                + Ví dụ tung đồng xu 5 lần, xác suất để 3 lần được mặt ngửa
                => n = 5; k = 3; p = 0.5
                => P(X=3) = (5!/3! x 2!) x 0.5^3 x (1 - 0.5)^5-3 = (5!/3! x 2!) x 0.5^3 x 0.5^2 = 10 x 0.125 x 0.25 = 0.3125 = 31.25%

        Phân phối Bernoulli (Bernoulli Distribution)
            - Áp dụng cho các sự kiện chỉ có thể có 2 biến cố và loại trừ nhau (mỗi lần thử chỉ có thể ra 1 biến cố) + thử 1 lần

            - Ví dụ: tung đồng xu 1 lần (tỷ lệ 50 - 50);  hệ thống AI nhận diện khuôn mặt đúng hay không (tỷ lệ khác 50 - 50)


        Phân phối Poisson (Poisson Distribution)
            - Phân phối Poisson mô hình hóa số lần xảy ra của các sự kiện với dữ liệu sự kiện rời rạc như số cuộc gọi nhận được mỗi giờ, số khách hàng đến mỗi ngày....

            - Điều kiện của một thí nghiệm Poisson:
                + Số sự kiện có thể đếm được.
                + Biết số lượng sự kiện trung bình xảy ra trong một khoảng thời gian.
                + Các sự kiện độc lập với nhau. 
            
            - Ví dụ: biết một trang web trung bình có 2 lượt xem mỗi giờ -> vẽ biểu đồ thể hiện số lượt xem trong 1 giờ bất kỳ -> khả năng có 1 lượt xem: cao; khả năng có 2 lượt xem: cao; khả năng có 3, 4, 5... lượt xem sẽ thấp dần

            - Công thức: 
                P(X=k)=λ^k x e^−λ​ / k!
                    X=k: Xác suất sự kiện xảy ra đúng k lần
                    λ(lambda): số lần trung bình xảy ra (ví dụ: trung bình 2 cuộc gọi 1 phút => λ = 2)
                    k: số lần mong muốn sự kiện xảy ra
                    e ~ 2.718: số Euler

                Ví dụ: Trung bình có 2 email mỗi phút -> λ. Tính xác suất có 3 email trong 1 phút -> k = 3
                    P(X=3) = (2^3 x 2.718 ^-3) / 3! = (8 x 0.1353) / 6 = 0.1804 ~ 18 %

Hàm Xác Suất và Mật Độ Xác Suất
    - Hàm khối xác suất - Probability Mass Function - PMF: dành cho biến rời rạc, ví dụ xác suất tung xúc sắc được số 3 P(X=3) = 1/6
        + Các phân phối dùng PMF: Bernoulli, Poisson, Binomial, Geometric
        
    - Hàm mật độ xác suất - Probability Density Function - PDF: dành cho biến liên tục, ví dụ chiều cao cây anh đào trung bình trong khoảng 14.5 - 15.5 feet
        + Các phân phối dùng PDF: Normal (Gaussian), Exponential, Uniform (continuous)

    - Phân phối chuẩn Gaussian: 
        + Là một phân phối xác suất liên tục có dạng đối xứng hình chuông (bell curve). Nó được gọi là phân phối Gauss, đặt theo tên nhà toán học Carl Gauss.

        + Đặc điểm quan trọng:
            . Dạng chuông đối xứng quanh giá trị trung bình.
            . Trung bình (Mean) nắm ở đỉnh của đồ thị.
            . 50% dữ liệu năm bên trái, 50% nam bên phải trung bình.
            . Tổng diện tích dưới đường cong = 1 (100% xắc suất).
            . Ví dụ: Chiều cao ngưoi truong thanh thuờng tuan theo phan phối chuẩn, voi đa số co chieu cao gan trung bình và rất ít người có chiều cao quá thấp hoặc quá cao.

        + Quy tắc thực nghiệm (Empirical Rule): theo quy tắc quy tắc 68-95-99.7
            . 68% giá trị nằm trong 1 độ lệch chuẩn (SD) của trung bình.
            . 95% giá trị nắm trong 2 độ lệch chuẩn.
            . 99.7% giá trị nằm trong 3 độ lệch chuẩn.
            . Ví dụ: Chiều cao cây anh đào (Mean = 15 feet, SD = 2 feet):
                . 68% cây có chiều cao 13-17 feet.
                . 95% cây có chiều cao 11-19 feet.
                . 99.7% cây có chiều cao 9-21 feet.
        
        + Z-Score
