```python
# load Data frame from csv

df_training = pd.read_csv('datab10.csv')
df_training.head()

# print all header
print(df_training.columns)

# infomation about data
print(df_training.info())

print(df_training.describe()) # mean: gia tri trung binh

# count null
print(df_training.isnull().sum())

# clean null records
df_training_clean = df_training.dropna()
print(len(df_training), len(df_training_clean))


# clean null basing on the specific columns => column cua 
df_training_clean = df_training.dropna(subset=['Calories'])
print(len(df_training), len(df_training_clean))

# fill null column with value
df_training_clean =  df_training.fillna(0, inplace=False)
print(df_training_clean.isnull().sum())

# fill with average value =
duration_mean = df_training['Duration'].mean()
print(duration_mean)
df_training_clean = df_training.copy()
df_training_clean['Duration'].fillna(duration_mean, inplace=True)
print(df_training_clean['Duration'].mean())

# fill with min value =
duration_min = df_training['Duration'].min()
print(duration_min)
df_training_clean = df_training.copy()
df_training_clean['Duration'].fillna(duration_min, inplace=True)
print(df_training_clean['Duration'].min())
```
