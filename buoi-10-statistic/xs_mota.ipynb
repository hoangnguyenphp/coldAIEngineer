```python
import pandas as pd

import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('USImport.csv', delimiter=';')
df.head()
# convert 'Quantity' to integer
df['Quantity'] = df['Quantity'].str.replace(',', '.').astype(float).astype(int)

df['Weight'] = df['Weight'].str.replace(',', '.').astype(float)

df['Value (US$)'] = df['Value (US$)'].str.replace(',', '.').astype(float)

df.info()

df.describe()


plt.figure(figsize=(10,6))
sns.histplot(df['Quantity'], bins=50, kde=True)
plt.title('Distribution of Quantity')
plt.xlabel('Quantity')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Weight'], bins=50, kde=True)
plt.title('Distribution of Weight')
plt.xlabel('Weight')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x=df['Weight'])
plt.title('Boxplot of Weight')
plt.xlabel('Weight')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Value (US$)'], bins=50, kde=True)
plt.title('Distribution of Value (US$)')
plt.xlabel('Value (US$)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x=df['Value (US$)'])
plt.title('Boxplot of Value (US$)')
plt.xlabel('Value (US$)')
plt.show()
```

```python
print(f'Mean of quantity: {df['Quantity'].mean():.2f}')
print(f'Median of quantity: {df['Quantity'].median():.2f}')
print(f'Mode of quantity: {df['Quantity'].mode()}')
print(f'Std of quantity: {df['Quantity'].std():.2f}')
print(f'CV of quantity (hệ số biến thiên): {(df['Quantity'].std() / df['Quantity'].mean()) * 100:.2f}%')
```

```python
print(f'Mean of Weight: {df['Weight'].mean():.2f}')
print(f'Median of Weight: {df['Weight'].median():.2f}')
print(f'Mode of Weight: {df['Weight'].mode()}')
print(f'Std of Weight: {df['Weight'].std():.2f}')
print(f'CV of Weight (hệ số biến thiên): {(df['Weight'].std() / df['Weight'].mean()) * 100:.2f}%')
```

```python
print(f'Mean of Value (US$): {df['Value (US$)'].mean():.2f}')
print(f'Median of Value (US$): {df['Value (US$)'].median():.2f}')
print(f'Mode of Value (US$): {df['Value (US$)'].mode()}')
print(f'Std of Value (US$): {df['Value (US$)'].std():.2f}')
print(f'CV of Value (US$) (hệ số biến thiên): {(df['Value (US$)'].std() / df['Value (US$)'].mean()) * 100:.2f}%')
```
