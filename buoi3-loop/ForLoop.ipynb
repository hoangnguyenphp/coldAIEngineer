```python
# range(n)
for i in range(10):
    print(i + 1, end=" ")
print()

# range(start, n)
for i in range(2, 10):
    print(i, end=" ")
print()

# range(start, n, step)

for i in range(1, 10, 2):
    print(i, end=" ")

print()


n = int(input("Enter a number:"))

for i in range(1, 11):
    print(f'{i:2} x {n} = {i*n:2}')


n = int(input("Enter a number:"))
for i in range(1, n + 1):
    for j in range(i):
        print("*", end="")
    print()
```
