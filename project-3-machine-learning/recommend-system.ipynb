```python
import pandas as pd
import numpy as np
```

```python

amazon_electronic_df = pd.read_csv('electronics.csv')
```

```python
amazon_electronic_df
```

```python
amazon_electronic_df.info()
```

```python
cols = ['item_id', 'user_id', 'rating', 'model_attr', 'category', 'brand', 'year']

amazon_electronic_eda_df = amazon_electronic_df[cols]

amazon_electronic_eda_df[cols] = (
    amazon_electronic_eda_df[cols]
        .replace(r'^\s*$', np.nan, regex=True)  # empty or spaces → NaN
        .fillna('Unknown')
)
```

```python
amazon_electronic_eda_df
```

```python
amazon_electronic_eda_df.info()
```

```python
n = 50000
step = len(amazon_electronic_eda_df)//n
print(step)

amazon_electronic_sample_df = amazon_electronic_eda_df[::step]
```
