Tổng Quan Về Khai Phá Luật Kết Hợp
    1. Định Nghĩa 
        Khai phá luật kết hợp (Association Rule Mining) là một kỹ thuật trong
        khai phá dữ liệu được sử dụng để tìm ra các mối quan hệ ẩn hoặc các
        mẫu quan trọng giữa các biến trong một tập dữ liệu lớn. Kỹ thuật này
        thường được áp dụng trong các lĩnh vực như kinh doanh bán lẻ, phân
        tích giỏ hàng, và hệ thống gợi ý.

    2. Khái Niệm Cơ Bản
        a. Luật Kết Hợp (Association Rule):
        · Luật kết hợp là một câu lệnh dưới dạng X -> Y, trong đó X và Y là
        các tập hợp con của tập mục (itemset)
        · Ví dụ: Trong một cửa hàng, nếu một khách hàng mua bánh mì (X),
        thì khả năng cao họ cũng mua bơ (Y)

        b. Tập Mục Phổ Biến (Frequent Itemset):
        · Tập mục phổ biến là một tập hợp các mục xuất hiện cùng nhau
        thuờng xuyen trong tap dữ liệu
        · Ví dụ: Tập mục {bánh mì, bơ} có thể là một tập mục phổ biến nếu
        nó xuất hiện trong nhiều giỏ hàng

        c. Độ Đo Chất Lượng Luật
        · Support (Hỗ trợ): Tỷ lệ phần trăm giao dịch trong tập dữ liệu chứa
        cả X và Y
        · Confidence (Độ tin cậy): Xác suất mà Y sẽ xuất hiện trong các giao
        dịch đã chứa X
        · Lift (Độ nâng): Đánh giá sự phụ thuộc giữa X và Y. Lift lớn hơn 1
        chỉ ra rằng X và Y xuất hiện cùng nhau thường xuyên hơn mong
        đợi nếu chúng độc lập.

    3. Thuật Toán Khai Phá Luật Kết Hợp
        a. Thuật Toán Apriori
        · Apriori là một trong những thuật toán cơ bản nhất được sử dụng để tìm tập mục phổ biến.
        · Nguyên lý Apriori: Mọi tập mục con của một tập mục phổ biến cũng phải là tập mục phổ biến.
        · Quy trình: Tìm tất cả các tập mục phổ biến, sau đó sử dụng chúng để tạo ra các luật kết hợp.

        b. Thuật Toán FP-Growth
        . FP-Growth (Frequent Pattern Growth) là một thuật toán thay thế cho Apriori, sử dụng cấu trúc dữ liệu gọi là FP-Tree để lưu trữ thông tin về tập mục phổ biến một cách hiệu quả
        . Ưu điểm: FP-Growth thường nhanh hơn Apriori vì nó không cần quét lại toàn bộ cơ sở dữ liệu nhiều lần

    4. Ứng Dụng Thực Tiễn
        a. Phân Tích Giỏ Hàng
        · Xác định các sản phẩm thường được mua cùng nhau để tối ưu
        hóa cách bày biện sản phẩm và chiến lược bán hàng

        b. Hệ Thống Gợi Ý
        · Đề xuất sản phẩm cho người dùng dựa trên các mẫu mua sắm
        trước đó của họ và các người dùng tương tự

        c. Phân Tích Du Liệu Y Tế
        · Tìm ra các mối liên hệ giữa các triệu chứng và bệnh tật để hỗ trợ chẩn đoán và điều trị


    5. Ví Dụ Minh Họa: Giả sử có tập dữ liệu về các giao dịch mua hàng tại một siêu thị

        Giao Dịch                   Mục Hàng              
                                        
        1                           Bánh Mì, Sữa

        2                           Bánh Mì, Bơ

        3                           Sữa, Bơ, Bánh Mì

        4                           Bánh Mì, Bơ, Sữa, Trứng

        5                           Sữa, Trứng


        => Từ dữ liệu trên, một luật kết hợp có thể là:
        · Luật: {Bánh mì} -> {Bơ}
        · Support: 3/5 = 60%
        · Confidence: 3/4 = 75%
        · Lift: (75%) / (3/5) = 1.25

Thuật Toán Apriori Trong Khai Phá Luật Kết Hợp

    Quy Trình Của Thuật Toán Apriori
        Bước 1: Tạo tập mục C1
        · Tạo danh sách các tập mục 1 mục (1-itemset) từ cơ sở dữ liệu
        giao dịch
        · Tính độ hỗ trợ (support) của từng tập mục

        Buớc 2: Lọc Tập Mục Pho Biến L1
        · Lọc ra các tập mục phổ biến từ C1, tạo thành tập L1, sao cho độ
        hỗ trợ của mỗi tập mục trong L1 lớn hơn hoặc bằng ngưỡng hỗ trợ
        tối thiểu

        Bước 3: Tạo Tập Mục Kế Tiếp Ck
        · Sử dụng tập mục phổ biến Lk từ bước trước để tạo ra các tập mục
        có k+1 mục (k-itemset)
        · Sử dụng nguyên lý Apriori để giảm số lượng tập mục cần kiểm tra:
        chỉ giữ lại các tập mục có tất cả các tập con k-mục đều phổ biến

        Buớc 4: Tinh Dộ Hỗ Tro va Loc Tap Muc Pho Bien Lk
        · Tính độ hỗ trợ (support) cho các tập mục trong Ck+
        · Lọc ra các tập mục phổ biến từ Ck+1 để tạo thành Lk+1

        Bước 5: Lặp Lại
        · Lặp lại các bước 3 và 4 cho đến khi không còn tập mục phổ biến
        nào nữa

        Bước 6: Tạo Luật Kết Hợp
        · Từ các tập mục phổ biến, tạo ra các luật kết hợp X->Y và tính độ
        tin cậy (confidence) của chúng
        · Chỉ giữ lại các luật có độ tin cậy lớn hơn hoặc bằng ngưỡng tin cậy
        tối thiểu

    Ví Dụ Minh Họa
        Giả sử có tập dữ liệu về các giao dịch mua hàng tại một siêu thị

        Giao Dịch                   Mục Hàng

            1                           Bánh Mì, Sữa

            2                           Bánh Mì, Bơ

            3                           Sữa, Bơ, Bánh Mì

            4                           Bánh Mì, Bơ, Sữa, Trứng

            5                           Sữa, Trứng

        Ngưỡng hỗ trợ tối thiểu (min_support): 2 giao dịch (40%)

        Bước 1: Tạo Tập Mục C1:
            · {Bánh mì}, {Sữa}, {Bơ}, {Trứng}

        Bước 2: Lọc Tập Mục Pho Biến L1
        · {Bánh mì} (4/5 = 80%)
        · {Sữa} (4/5 = 80%)
        · {Bơ} (3/5= 60%)
        · {Trứng} (2/5 = 40%)

        Bước 3: Tạo Tập Mục Kế Tiếp C2
        · {Bánh mì, Sữa}, {Bánh mì, Bơ}, {Bánh mì, Trứng}, {Sữa, Bơ},
        {Sữa, Trứng}, {Bơ, Trứng}

        Buoc 4: Tinh Dộ Hỗ Tro va Loc Tap Muc Pho Bien L2
        · {Bánh mì, Sữa} (3/5 = 60%)
        · {Bánh mì, Bơ} (3/5 = 60%)
        · {Sữa, Bơ} (2/5 = 40%)
        · {Sữa, Trứng} (2/5 = 40%)

        Bước 5: Tạo Tập Mục Kế Tiếp C3
        · {Bánh mì, Sữa, Bơ}

        Buoc 6: Tinh Độ Hỗ Tro va Loc Tap Muc Pho Bien L3
        · {Bánh mì, Sữa, Bơ} (2/5 = 40%)

        Bước 7: Tạo Luật Kết Hợp
        · {Bánh mì} -> {Sữa} (Độ tin cậy = 3/4 = 75%)
        · {Sữa} - {Bánh mì} (Độ tin cậy = 3/4 = 75%)
        · {Bánh mì} -+ {Bơ} (Độ tin cậy = 3/4 = 75%)
        · {Bơ} -+ {Bánh mì} (Độ tin cậy = 3/3 = 100%)

        Từ tập mục phổ biến L3, chúng ta có thể tạo ra luật kết hợp như sau:
            · Luật: {Bánh mì} -+ {Sữa, Bơ}
            o Độ hỗ trợ (Support) của {Bánh mì, Sữa, Bơ}: 2/5 = 40%
            o Độ tin cậy (Confidence) = Số giao dịch chứa cả {Bánh mì,
            Sữa, Bơ} / Số giao dịch chứa {Bánh mì} = 2/4 = 50%

            Ngoài luật kết hợp đã có ở trên, chúng ta cũng có thể có
            · Luật: {Bánh mì} -> {Sữa, Bơ} (Độ tin cậy = 2/4 = 50%)
            · Luật: {Sữa} -+ {Bánh mì, Bơ} (Độ tin cậy = 2/4 = 50%)
            · Luật: {Bơ} - {Bánh mì, Sữa} (Độ tin cậy = 2/3 = 66.7%)

Thuật Toán FP-Growth Trong Khai Phá Luật Kết Hợp

    1. Giới Thiệu
    Thuật toán FP-Growth (Frequent Pattern Growth) là một phương pháp
    hiệu quả để khai phá các tập mục phổ biến (frequent itemsets) trong
    các cơ sở dữ liệu lớn. Khác với thuật toán Apriori, FP-Growth không
    cần phải quét lại toàn bộ cơ sở dữ liệu nhiều lần, giúp tiết kiệm thời
    gian và tài nguyên tính toán.

    2. Các Khái Niệm Cơ Bản
    a. Itemset (Tập mục): Một tập hợp các mục trong một cơ sở dữ liệu
    giao dịch
    b. Support (Độ hỗ trợ): Tỷ lệ phần trăm giao dịch trong cơ sở dữ liệu
    chứa một tập mục nào đó
    c. Frequent itemset (Tập mục phổ biến): Tập mục có độ hỗ trợ lớn
    hơn hoặc bằng ngưỡng hỗ trợ tối thiểu (min_support)
    d. FP-Tree (Cây mẫu phổ biến): Một cấu trúc dữ liệu cây được sử
    dụng để lưu trữ thông tin về các tập mục phổ biến trong cơ sở dữ liệu.

    3. Quy Trình Của Thuật Toan FP-Growth

Độ Độ Lift Trong Khai Phá Luật Kết Hợp



```python
import pandas as pd
import numpy as np
from apyori import apriori

df_retail = pd.read_csv('Online_Retail_5000.csv')
df_retail.head()

df_retail['StockCode'].nunique()


df_retail['InvoiceNo'] = df_retail['InvoiceNo'].astype('str') + ' '
df_retail['StockCode'] = df_retail['StockCode'].astype('str') + ' '

df_retail['InvoiceNo'].head()

transactions = []
invoices = df_retail['InvoiceNo'].unique()
print(len(invoices))

for iv in invoices:
    items = df_retail[df_retail['InvoiceNo'] == iv]['StockCode'].tolist()
    transactions.append(items)

print(len(transactions))

print(transactions[0:10])

```

```python
apr_rules = apriori(transactions, min_support=0.02, min_confidence=0.3, min_lift=3, max_length=5)
apr_rules = list(apr_rules)

apr_rules_sorted = sorted(apr_rules, key=lambda x: x.support, reverse=True)
for i, rule in enumerate(apr_rules_sorted[:5]):
    print(f"Rule {i+1}:")
    print(f"Items: {rule.items}")
    print(f"Support: {rule.support}")
    for ordered_stat in rule.ordered_statistics:
        print(f'  {list(ordered_stat.items_base)} -> {list(ordered_stat.items_add)}')
        print(f"  Confidence: {ordered_stat.confidence}")
        print(f"  Lift: {ordered_stat.lift}")
    print("=====================================")
```
