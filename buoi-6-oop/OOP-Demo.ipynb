```python
class Student:
    # Constructor
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def hi(self):
        print(f'My name is {self.name}. I am {self.age}  years old.')

john = Student("Jonh", 20)

print(john.name)
print(john.age)
john.hi()

john.age = 10
john.grade = 10
print(john.name)
print(john.age)
print(john.grade)
```

```python
# private public
class Student:
    # Constructor
    def __init__(self, name, age):
        self.name = name # name is public
        self.__age = age # age is private

    def hi(self):
        print(f'My name is {self.name}. I am {self.__age}  years old.')

john = Student("Jonh", 20)

print(john.__age) # => AttributeError: 'Student' object has no attribute '__age'
```

```python
# getter setter

class Student:
    # Constructor
    def __init__(self, name, age):
        self.name = name # name is public
        self.setAge(age) # age is private

    def hi(self):
        print(f'My name is {self.name}. I am {self.__age}  years old.')

    def getAge(self):
        return self.__age
    
    def setAge(self, newAge):
        if newAge >= 18:
            self.__age = newAge
        else:
            print("Age cannot be under 18.")

john = Student("Jonh", 18)
john.hi()
# print(john.__age) # => AttributeError    
john.__age = 16 # no exception but no assigning action

john.setAge(17)
```

```python
# getter setter

class Student:
    # Constructor
    def __init__(self, name, p_age):
        self.name = name # name is public
        self.__age = p_age

    def hi(self):
        print(f'My name is {self.name}. I am {self.__age}  years old.')

john = Student("Jonh", 18)
print(john.hi())

john.age = 19
print(john.age)
print(john.hi())
```

```python
# Property

class Student:
    # Constructor
    def __init__(self, name, age):
        self.name = name # name is public
        self.age = age

    def hi(self):
        print(f'My name is {self.name}. I am {self.__age}  years old.')

    @property
    def age(self):
        return self.__age
    
    @age.setter
    def age(self, newAge):
        if newAge > 18:
            self.__age = newAge
        else:
            print("Age cannot be under 18.")


michael = Student("Michael", 20)
print(michael.name)
print(michael.age)
michael.hi()

michael.age = 16
print(michael.name)
print(michael.age)

michael.age = 25
print(michael.name)
print(michael.age)
```

```python
# Inheritance

class Person:
    def __init__(self, name):
        self.name = name

    @property
    def name(self):
        return self.__name

    @name.setter
    def name(self, name):
        if name != '':
            self.__name = name
        else:
            print('Name cannot be empty.')

class Student(Person):
    def __init__(self, name, grade):
        super().__init__(name)
        self.grade = grade
    
    @property
    def grade(self):
        return self.__grade
    
    @grade.setter
    def grade(self, grade):
        self.__grade = grade

david = Student('David', 20)        
print(david.name)
print(david.grade)
```

```python
# Override
class Person:
    def __init__(self, name):
        self.name = name

    @property
    def name(self):
        return self.__name

    @name.setter
    def name(self, name):
        if name != '':
            self.__name = name
        else:
            print('Name cannot be empty.')

    def hello(self):
        print(self.name)

class Student(Person):
    def __init__(self, name, grade):
        super().__init__(name)
        self.grade = grade
    
    @property
    def grade(self):
        return self.__grade
    
    @grade.setter
    def grade(self, grade):
        self.__grade = grade

    def hello(self):
        super().hello()
        print(self.grade)

david = Student('David', 20)        
david.hello()
```

```python
class Employee:
    BASIC_SALARY = 1000

    def __init__(self, name, grade):
        self.name = name
        self.grade = grade

    @property
    def name(self):
        return self.__name
    
    @name.setter
    def name(self, name):
        self.__name = name

    @property
    def grade(self):
        return self.__grade
    
    @grade.setter
    def grade(self, grade):
        self.__grade = grade

    def salary(self):
        return self.grade * self.BASIC_SALARY
    

jane = Employee('Jane', 3)
jane.BASIC_SALARY = 50

print(jane.salary())
```
