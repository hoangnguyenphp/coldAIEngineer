Bài 1: Tạo lớp SinhVien
Tạo một lớp SinhVien với các thuộc tính ten, tuoi, và phương thức
gioi_thieu() để in ra thông tin sinh viên.

```python
class SinhVien:
    def __init__(self, ten, tuoi):
        self.ten = ten
        self.tuoi = tuoi

    @property
    def ten(self):
        return self.__ten
    
    @ten.setter
    def ten(self, ten):
        self.__ten = ten

    def gioi_thieu(self):
        print(f'I am {self.ten}, I am {self.tuoi} year olds.')

duong = SinhVien("Duong", 20)
print(duong.ten)
duong.gioi_thieu()
```

Bài 2: Tạo lớp HinhChuNhat

Tạo lớp HinhChuNhat với các thuộc tính chieu_dai và chieu_rong, và phương
thức dien_tich() để tính diện tích hình chữ nhật.

```python
class HinhChuNhat:
    def __init__(self, chieuDai, chieuRong):
        self.chieuDai = chieuDai
        self.chieuRong = chieuRong

    @property
    def chieuDai(self):
        return self.__chieuDai
    
    @chieuDai.setter
    def chieuDai(self, chieuDai):
        self.__chieuDai = chieuDai

    @property
    def chieuRong(self):
        return self.__chieuRong
    
    @chieuRong.setter
    def chieuRong(self, chieuRong):
        self.__chieuRong = chieuRong

    def dien_tich(self):
        return self.chieuDai * self.chieuRong
    

hcn1 = HinhChuNhat(20, 10)
print(f'Dien tich hinh chu nhat: {hcn1.dien_tich()}')
```

Bài 3: Thêm phương thức vào lớp
Thêm phương thức chu_vi() vào lớp HinhChuNhat để tính chu vi hình chữ
nhật.

```python
class HinhChuNhat:
    def __init__(self, chieuDai, chieuRong):
        self.chieuDai = chieuDai
        self.chieuRong = chieuRong

    @property
    def chieuDai(self):
        return self.__chieuDai
    
    @chieuDai.setter
    def chieuDai(self, chieuDai):
        self.__chieuDai = chieuDai

    @property
    def chieuRong(self):
        return self.__chieuRong
    
    @chieuRong.setter
    def chieuRong(self, chieuRong):
        self.__chieuRong = chieuRong

    def dien_tich(self):
        return self.chieuDai * self.chieuRong
    
    def chu_vi(self):
        return (self.chieuDai + self.chieuRong) * 2
    

hcn1 = HinhChuNhat(20, 10)
print(f'Dien tich hinh chu nhat: {hcn1.dien_tich()}')
print(f'Chu vi hinh chu nhat: {hcn1.chu_vi()}')
```

Bài 4: Thay đổi thuộc tính

Viết chương trình để thay đổi giá trị của thuộc tính tuoi trong lớp
SinhVien.

```python
class SinhVien:
    def __init__(self, ten, tuoi):
        self.ten = ten
        self.tuoi = tuoi

    @property
    def ten(self):
        return self.__ten
    
    @ten.setter
    def ten(self, ten):
        self.__ten = ten

    @property
    def tuoi(self):
        return self.__tuoi
    
    @tuoi.setter
    def tuoi(self, tuoi):
        self.__tuoi = tuoi

    def gioi_thieu(self):
        print(f'I am {self.ten}, I am {self.tuoi} year olds.')

duong = SinhVien("Duong", 20)
duong.gioi_thieu()

duong.tuoi = 18
duong.gioi_thieu()
```

Bài 5: Kiểm tra thuộc tính riêng tư

Tạo lớp HocSinh với thuộc tính riêng tư diem, sau đó sử dụng phương thức
getter và setter để truy cập và thay đổi điểm.

```python
class HocSinh:
    def __init__(self, ten, diem):
        self.__ten = ten
        self.__diem = diem

    def getTen(self):
        return self.__ten

    def setTen(self, ten):
        self.__ten = ten

    def getDiem(self):
        return self.__diem
    
    def setDiem(self, diem):
        self.__diem = diem

duong = HocSinh("Duong", 10)
print(f'Diem cua ban: {duong.getDiem()}')

duong.setDiem(8)
print(f'Diem cua ban: {duong.getDiem()}')
```

Bài 6: Kế thừa lớp
Tạo một lớp GiaoVien kế thừa từ lớp Nguoi và thêm phương thức day_hoc().

```python
class Nguoi:
    def __init__(self, ten):
        self.ten = ten

    @property
    def ten(self):
        return self.__ten
    
    @ten.setter
    def ten(self, ten):
        self.__ten = ten

class GiaoVien(Nguoi):

    def day_hoc(self):
        print("Dang day hoc.")

john = GiaoVien("John")

print(john.ten)
print(john.day_hoc())
```

Bài 7: Ghi đè phương thức

Tạo lớp GiaoVien kế thừa từ lớp Nguoi, và ghi đè phương thức gioi_thieu() để
in ra thông tin về giáo viên.

```python
class Nguoi:
    def __init__(self, ten):
        self.ten = ten

    @property
    def ten(self):
        return self.__ten
    
    @ten.setter
    def ten(self, ten):
        self.__ten = ten

    def gioi_thieu(self):
        print(self.ten)

class GiaoVien(Nguoi):

    def day_hoc(self):
        print("Dang day hoc.")

    def gioi_thieu(self):
        print(f'Ten giao vien: {self.ten}')

john = GiaoVien("John")

print(john.ten)
print(john.day_hoc())
john.gioi_thieu()
```

Bài 8: Định nghĩa lớp PhuongTien
Tạo lớp PhuongTien với thuộc tính loai_phuong_tien và phương
thức di_chuyen() để in ra loại phương tiện và cách di chuyển.

```python
class PhuongTien:
    def __init__(self, loaiPhuongTien):
        self.loaiPhuongTien = loaiPhuongTien

    @property
    def loaiPhuongTien(self):
        return self.__loaiPhuongTien
    
    @loaiPhuongTien.setter
    def loaiPhuongTien(self, loaiPhuongTien):
        self.__loaiPhuongTien = loaiPhuongTien

    def di_chuyen(self):
        print(f"{self.loaiPhuongTien} di chuyen.")

phuongTien = PhuongTien("xe may")
phuongTien.di_chuyen()
```

Bài 9: Tạo lớp XeMay và OTo

Tạo hai lớp XeMay và OTo kế thừa từ lớp PhuongTien. Thêm các thuộc
tính và phương thức riêng như toc_do và so_ghe.

```python
class PhuongTien:
    def __init__(self, loaiPhuongTien):
        self.loaiPhuongTien = loaiPhuongTien

    @property
    def loaiPhuongTien(self):
        return self.__loaiPhuongTien
    
    @loaiPhuongTien.setter
    def loaiPhuongTien(self, loaiPhuongTien):
        self.__loaiPhuongTien = loaiPhuongTien

    def di_chuyen(self):
        print(f"{self.loaiPhuongTien} di chuyen.")


class XeMay(PhuongTien):
    def __init__(self, loaiPhuongTien, tocDo, soGhe):
        super().__init__(loaiPhuongTien)
        self.tocDo = tocDo
        self.soGhe = soGhe

    @property
    def tocDo(self):
        return self.__tocDo
    
    @tocDo.setter
    def tocDo(self, tocDo):
        self.__tocDo = tocDo

    @property
    def soGhe(self):
        return self.__soGhe
    
    @soGhe.setter
    def soGhe(self, soGhe):
        self.__soGhe = soGhe         

class Oto(PhuongTien):
    def __init__(self, loaiPhuongTien, tocDo, soGhe):
        super().__init__(loaiPhuongTien)
        self.tocDo = tocDo
        self.soGhe = soGhe

    @property
    def tocDo(self):
        return self.__tocDo
    
    @tocDo.setter
    def tocDo(self, tocDo):
        self.__tocDo = tocDo

    @property
    def soGhe(self):
        return self.__soGhe
    
    @soGhe.setter
    def soGhe(self, soGhe):
        self.__soGhe = soGhe   


xemay = XeMay("xe may", 50, 2)
xemay.di_chuyen()
print(f'Toc do: {xemay.tocDo}')
print(f'So ghe: {xemay.soGhe}')

oto = Oto("oto", 100, 20)
oto.di_chuyen()
print(f'Toc do: {oto.tocDo}')
print(f'So ghe: {oto.soGhe}')
```
