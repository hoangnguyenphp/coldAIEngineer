```python
import pandas as pd
animals = pd.read_csv('du_lieu_dong_vat.csv')
animals.head()
```

```python
documents = animals['text'].tolist()
documents
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(documents)
```

```python
print(X.shape)  # in ra kích thước ma trận TF-IDF
```

```python
names = vectorizer.get_feature_names_out()
print(names[100:150])  
```

```python
print(X[0])
```

```python
model = KMeans(n_clusters=3, random_state=42, init='k-means++')
model.fit(X)
```

```python
centroids = model.cluster_centers_.argsort()[:, ::-1]

for i in range(3):
    print(f"Cluster {i}:")
    for ind in centroids[i, :5]:  # in ra 5 từ hàng đầu cho mỗi cụm
        print(f" {names[ind]}")
```

```python
a_text = 'Poodles are highly intelligent, active, and hypoallergenic dogs known for their dense, curly coats and versatile, athletic builds. Available in Standard, Miniature, and Toy sizes, they are affectionate, easily trained, and require regular, professional grooming. Originally water retrievers, they are playful, energetic companions that excel in dog sports.'
```

```python
target = vectorizer.transform([a_text])
prediction = model.predict(target)
print(f'The input text belongs to cluster: {prediction[0]}')
```
