```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

n_employee = 50
salary_std = 3 # σ = sigma
salary_mean = 18
salary_samples = np.random.normal(size=n_employee, scale=salary_std, loc=salary_mean) # loc = μ = mu = Giá trị trung bình (mean)

plt.hist(salary_samples, bins=10, edgecolor='black')
plt.title('Employee Salary Distribution')
plt.xlabel('Salary (in thousands)')
plt.ylabel('Number of Employees')
plt.show()


```

```python
# Công thức Khoảng tin cậy = X(mean) +- (Z  x (σ/√n)) khi n >= 30 hay còn gọi là phân phối Z, phân phối chuẩn (z)

n_employee = 50
salary_std = 3 # σ = sigma
salary_mean = 18
salary_samples = np.random.normal(size=n_employee, scale=salary_std, loc=salary_mean) # loc = μ = mu = Giá trị trung bình (mean)

confidence = 0.95
alpha = 1 - confidence
z_score = stats.norm.ppf(1 - alpha/2) # Z (ppf: Percent Point Function => Chính là hàm nghịch đảo của CDF)
print(z_score)
```

```python

margin = z_score * salary_std / n_employee ** 0.5 ## n_employee ** 0.5 => căn bậc 2 của n_employee
ci_lower = salary_mean - margin
ci_upper = salary_mean + margin
print(f"95% Confidence Interval for the mean salary: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
# Công thức Khoảng tin cậy = X(mean) +- (t  x (s/√n)) khi n < 30 hay còn goi là phân phối t-Student (t)
n_workers = 20 # n
times = np.random.normal(size=n_workers, scale=2, loc=10)
time_mean = np.mean(times) # X
time_std = np.std(times, ddof=1) # s

confidence = 0.95
alpha = 1 - confidence
t_score = stats.t.ppf(1 - alpha/2, df=n_workers - 1) # t
print(f't score: {t_score}')
margin = t_score * time_std / n_workers ** 0.5
ci_lower = time_mean - margin # ci: confident interval
ci_upper = time_mean + margin # ci: confident interval
print(f"Mean time taken: {time_mean:.2f} minutes")
print(f"Standard Deviation of time taken: {time_std:.2f} minutes")
print(f"95% Confidence Interval for the mean time: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
# Ước lượng tỷ lệ tổng thể: Khoảng tin cậy: p +- Z x √(p x (1-p))/n
n_customers = 200 # n
n_customers_happy = 160
p_hat = n_customers_happy / n_customers # p 
print(f"Sample Proportion of Happy Customers: {p_hat:.2f}")
confidence = 0.99
alpha = 1 - confidence               # 0.01
z_score = stats.norm.ppf(1 - alpha/2)
print(z_score)
margin = z_score * ((p_hat * (1 - p_hat)) / n_customers) ** 0.5
ci_lower = p_hat - margin # ci: confident interval
ci_upper = p_hat + margin # ci: confident interval
print(f"99% Confidence Interval for the proportion of happy customers: ({ci_lower:.2f}, {ci_upper:.2f})")
```

Kiểm định giả thuyết (Hypothesis Testing)
    1. Kiểm định giả thuyết (Hypothesis Testing)
    Kiểm định giả thuyết là một phương pháp thống kê nhằm đưa ra kết
    luận về một tham số của tổng thể dựa trên dữ liệu mẫu. Nó giúp xác
    định liệu một kết quả có thể xảy ra do ngẫu nhiên hay không.
    · Giả thuyết không (H0): Giả định ban đầu (ví dụ: không có sự khác
    biệt).
    · Giả thuyết đối (H1): Giả định thay thế (ví dụ: có sự khác biệt).


    2. Các bước cơ bản

    B1. Xác định giả thuyết:
    · H0: Không có khác biệt hoặc hiệu ứng (ví dụ: u = 100).
    · H1: Có khác biệt hoặc hiệu ứng (ví dụ: u # 100).
    B2. Chọn mức ý nghĩa (a):
    · Thường chọn a = 0.05 (tương ứng với 5% xác suất bác bỏ H0 sai lầm)
    B3. Tính toán thống kê kiểm định:
    · Sử dụng các công cụ như z-test, t-test, hoặc x2-test tùy vào loại dữ liệu

    B4. So sánh giá trị p (p-value):
        Nếu p ≤ a: Bác bỏ H0 (kết quả có ý nghĩa thống kê).
        Nếu p > a: Không đủ bằng chứng bác bỏ H0.
    B5. Đưa ra kết luận:
        Dựa trên kết quả kiểm định để trả lời câu hỏi nghiên cứu.

3. P-value: Định nghĩa và cách tính
    P-value là xác suất để quan sát được một giá trị kiểm định bằng hoặc
    cực đoan hơn giá trị thực tế, nếu giả thuyết gốc (H0) đúng.
    
    Công thức p-value:
    · Kiểm định 2 phía
        P-value = 2 · P(Thống kê ≥ |Giá trị thực tế|| H0)

    · Kiểm định 1 phía
         P-value = P(Thống kê ≥ Giá trị thực tế| H0)

        Hoặc: P-value = P(Thống kê ≤ Giá trị thực tế| H0)



```python
# t-test = (X(mean) - μ0 )/(s/√n)
# X: trung bình mẫu; μ0: trung bình giả định; s: độ lệch chuẩn mẫu; n: kích thước mẫu.

mean_declare = 7.5 # μ0

n_students = 25 # n
math_scores = np.random.normal(size=n_students, scale=1.0, loc=7.8)
mean_math = np.mean(math_scores)  # X
std_math = np.std(math_scores, ddof=1) # s
print (f"Mean Math Score: {mean_math :.2f}")
print(f"Standard Deviation of Math Scores: {std_math :.2f}")

alpha = 0.05 # a
t = (mean_math - mean_declare) / (std_math / n_students ** 0.5)

# P-value = 2 · P(Thống kê ≥ |Giá trị thực tế|| Họ)
#  => 0.5023 >  0.05 
# => p > a: Không đủ bằng chứng bác bỏ H0.
p_value = 2 * (1 - stats.t.cdf(abs(t), df=n_students -1)) 

print(f"T-statistic: {t :.2f}")
print(f"P-value: {p_value :.4f}")
```

```python
n_workers = 20
n_products = 75 # declare >= 75 products

products = np.array([78, 74, 80, 73, 77, 76, 79, 81, 72, 75, 74, 78, 77, 76, 80, 79, 73, 75, 78, 77])
mean_products = np.mean(products)
std_products = np.std(products, ddof=1)
print(f"Mean Products Produced: {mean_products:.2f}")
print(f"Standard Deviation of Products Produced: {std_products:.2f}")

alpha = 0.05
# H0: mean_products <= 75
# H1: mean_products > 75
t = (mean_products - n_products) / (std_products / n_workers ** 0.5)
p_value = 1 - stats.t.cdf(t, df=n_workers - 1)
print(f"T-statistic: {t:.2f}")
print(f"P-value: {p_value:.4f}")
if p_value <= alpha:
    print('Có bằng chứng cho thấy sản phẩm trung bình <= 75 (bác bỏ H0)')
else:
    print('Không có bằng chứng cho thấy sản phẩm trung bình <= 75 (không bác bỏ H0)')
```

```python
# z-test = (X(mean) - μ0 )/(σ/√n)
# X: trung bình mẫu; μ0: trung bình giả định; σ: độ lệch chuẩn tổng thể; n: kích thước mẫu.

weight = 500 # μ0
std_weight = 15 # σ

n_samples = 100 # n

weights = np.random.normal(size=n_samples, scale=std_weight, loc=500)
print(type(weights))
mean_weight = np.mean(weights) # X
std_weight_sample = np.std(weights, ddof=1)
print(f"Mean Weight: {mean_weight:.2f}")
print(f"Standard Deviation of Weights: {std_weight_sample:.2f}")

# H0: mean_weight < 500
# H1: mean_weight >= 500
z = (mean_weight - weight) / (std_weight_sample / n_samples ** 0.5)
print(z)
p_value = 1 - stats.norm.cdf(z)

alpha = 0.05
z_statistic = stats.norm.ppf(1 - alpha)
print(f"Z-statistic: {z:.2f}")
print(f"P-value: {p_value:.4f}")

if p_value <= alpha:
    print('Có bằng chứng cho thấy trọng lượng trung bình >= 500 (bác bỏ H0)')
else:
    print('Không có bằng chứng cho thấy trọng lượng trung bình >= 500 (không bác bỏ H0)')
```
