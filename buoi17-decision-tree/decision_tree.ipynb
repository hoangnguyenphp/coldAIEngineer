```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn. tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score
```

```python
df = pd.read_csv('data_decision_tree_bt.csv')
df.info()
```

```python
print(df['DoHaiLong'].unique())
print(df ['LoaiKhachHang'].unique())
print(df['LoaiHinh'].unique())
```

```python
df.isnull().sum()
```

```python
df_clean = df.dropna()
df_clean.isnull().sum()
```

```python
df_clean_encode = pd.get_dummies(df_clean, drop_first=True, dtype=int)
df_clean_encode.columns
df_clean_encode.head()
```

```python
# df_clean_encode = df_clean_encode.rename(columns={'DoHaiLong_Không Hai Lòng': 'DoHaiLong',
#                                                          'LoaiKhachHang_KH Trung Thanh': 'LoaiKhachHang',
#                                                          'LoaiHinh_Di Cung Gia Dình/Bạn Bè': 'LoaiHinh'})

df_clean_encode = df_clean_encode.rename(columns={'DoHaiLong_Không Hai Lòng': 'DoHaiLong',
                                                         'LoaiKhachHang_KH Trung Thanh': 'LoaiKhachHang',
                                                         'LoaiHinh_Di Cung Gia Dình/Bạn Bè': 'LoaiHinh'})
df_clean_encode.head()
```

```python
y = df_clean_encode['DoHaiLong']
x = df_clean_encode.drop(columns=['DoHaiLong', 'TT'])
print(x.shape(), y.shape())

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
tree = DecisionTreeClassifier(max_depth=4, random_state=42)
tree.fit(x_train, y_train)

tree_pred = tree.predict(x_test)
accuracy = accuracy_score(y_test, tree_pred)
f1_score = f1_score(y_test, tree_pred)
recall = recall_score(y_test, tree_pred)
precision = precision_score(y_test, tree_pred)
print(f'Accuracy: {accuracy}')
print(f'F1 Score: {f1_score}')

print(f'Recall: {recall}')
print(f'Precision: {precision}')
```
