Học Máy Là Gì va Gom Những Lĩnh Vực Nao?
    Học máy (Machine Learning) là một nhánh của trí tuệ nhân tạo (AI), tập trung vào xây dựng các hệ thống có khả năng tự học và cải thiện hiệu suất từ dữ liệu mà không cần lập trình cụ thể cho từng nhiệm vụ.

    Học Máy:

        Supervised Learning (Học có giám sát): Học từ dữ liệu đã được gán nhãn. Ví dụ: Dự đoán giá nhà.

        Unsupervised Learning (Học không Giám sát): Tìm cấu trúc ẩn trong dữ liệu chưa gán nhãn. Ví dụ: Phân nhóm khách hàng.

        Deep Learning (Học sâu): Sử dụng mạng nơ-ron đa lớp cho dữ liệu phức tạp. Ví dụ: Nhận diện khuôn mặt.

        Reinforcement Learning: Học qua tương tác và phản hồi từ môi trường. Ví dụ: Huấn luyện robot chơi cờ.

Đi Sâu vào Supervised Learning: Hai Bài Toán Cốt lõi
    Supervised Learning được chia thành 2 loại chính dựa trên đặc tính của biến mục tiêu (target variable): giá trị liên tục hoặc nhãn rời rạc:
        Regression (Hồi quy) Dữ liệu đầu ra: Liên tục

        Classification (Phân lớp) Dữ liệu đầu ra: Rời rạc

        => Lựa chọn phương pháp phụ thuộc vào câu hỏi bạn muốn trả lời: "Bao nhiêu?' (Regression) hay 'Loại nào?' (Classification).

            Bài Toán Hồi Quy (Regression): 
                Các thuật toán:
                    Linear Regression
                    SVR
                    Ridge Regression
                    Lasso Regression

                Ứng dụng Dự Báo Các Giá Trị Liên Tục
                    Đầu ra: Một giá trị số thực (ví dụ: giá nhà $150,000, nhiệt độ 25.5°C).
                    Mục tiêu: Dự đoán một con số.
                    Thước đo phổ biến: RMSE, MAE, R-squared.

            Bài Toán Phân Lớp (Classification): 
                Các thuật toán:
                    Logistic Regression
                    SVM
                    KNN
                    Decision Tree
                    Random Forest     
            
                Ứng dụng Dự Báo Các Nhãn Rời Rạc
                    Đầu ra: Một nhãn/lớp trong danh sách hữu hạn (ví dụ: spam/không spam, bệnh A/B/C).
                    Mục tiêu: Gán một đối tượng vào một nhóm.
                    Thước đo phổ biến: Accuracy, Precision, Recall, F1-Score.

                
                Decision Tree

                    Định nghĩa: Một mô hình có cấu trúc dạng cây, sử dụng các nút quyết định để phân loại dữ dữ liệu.

                    Thành phần chính:
                    · Root Node (Nút gốc): Điểm bắt đầu.
                    · Internal Nodes (Nút trung gian): Các câu hỏi/điều kiện.
                    · Leaf Nodes (Nút lá): Kết quả dự đoán cuối cùng.

                    Ưu điểm
                        Dễ hiểu, dễ giải thích: "Hộp trắng' (White box), có thể trình bày cho người không chuyên.
                        Không cần chuẩn hóa dữ liệu: Xử lý được nhiều loại dữ liệu.
                        Tự động chọn đặc trưng: Các đặc trưng ở gần gốc là quan trọng nhất.

                    Nhược điểm
                        Dễ bị overfitting: Cây quá sâu sẽ học thuộc lòng dữ liệu huấn luyện.
                        Không ổn định: Thay đổi nhỏ trong dữ liệu có thể tạo ra cây hoàn toàn khác.
                        Thiết vò với biến liên tục: Tạo ra các ngưỡng phân chia cứng nhắc.

                    Cây Quyết Định Được Xây Dựng Như Thế Nào: Nguyên tắc "Chia để trị" (Divide and Conquer): Tại mỗi nút, thuật toán chọn ra đặc trưng và ngưỡng phân chia tốt nhất để tạo ra các nhóm con "tinh khiết" nhất có thể.

                        Gini Impurity (Độ tạp chất Gini): Đo mức độ "lẫn lộn" của các lớp trong một nút. Mục tiêu: Giảm Gini.

                        Information Gain (Lợi ích thông tin): Đo sự giảm thiểu "hỗn loạn" (Entropy) sau khi phân chia. Mục tiêu: Tăng Information Gain.

                        Các bước:
                            Bắt đầu tại nút gốc. 
                            Tìm đặc trưng/ngưỡng phân chia tốt nhất (Gini/information Gain).
                            Chia dữ liệu thành các nút con. 
                            Lặp lại cho đến khi đạt điều kiện dừng.

                        Lớp là gì, nút là gì?

                    Đo Lường Sự "Tinh Khiết" với Chỉ Số Gini
                        Công thức: Gini = 1 - Σ(pi)^2 
                            * p, là tỷ lệ mẫu thuộc lớp i trong nút.
                        
                        Ví dụ 1 nút có 2 lớp: A chiếm 60%, B chiếm 40%
                            p(A) = 60/100 = 0.6
                            p(B) = 40/100 = 0.4
                            Gini = 1 - (0.62 + 0.42)
                            Gini = 1 - (0.36 + 0.16) = 1 - 0.52
                            => Gini = 0.48

                        Diễn giải: 
                            Gini = 0 có nghĩa là nút hoàn toàn tinh khiết (chỉ chứa 1 lớp)
                            Gini = 0.5 là độ tạp chất tố đa (trong trường hợp 2 lớp).

                    Tinh Chỉnh Siêu Tham Số: Chìa Khoa Chong Overfitting Một cây quyết định được "mọc" tự do sẽ trở nên quá phức tạp và học thuộc lòng dữ liệu huấn luyện (overfitting). Siêu tham số (hyperparameters) là các "luật" chúng ta đặt ra trước khi huấn luyện để kiểm soát sự phát triển của cây.

                        Overfitting học thuộc lòng, dự đoán kém trên dữ liệu mới.

                        Good Fit tổng quát hóa tốt.

                        Hyperparameters:  
                            max_depth: Ý nghĩa: Giới hạn số tầng tối đa của cây, từ gốc đến lá. Đây là cách trực tiếp nhất để kiểm soát độ phức tạp của mô hình.

                            min_samples_split (áp dụng trước khi chia): Ý nghĩa: Là số mẫu tối thiểu cần có trong một nút để nút đó được phép phân chia tiếp. Nếu số mẫu ít hơn, nó sẽ trở thành nút lá.

                            min_samples_leaf (áp dụng khi chia nếu số mẫu không đủ => hủy lượt chia): Là số mẫu tối thiểu cần có trong một nút lá. Một lượt phân chia chỉ hợp lệ nếu mỗi nút con sau khi chia đều có ít nhất `min_samples_leaf' mẫu.

    

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score

df = pd.read_csv('data_decision_tree_bt.csv')
df.head()

df_clean = df.dropna()
df_clean.isnull().sum()

df_cleaned_encoded = pd.get_dummies(df_clean, drop_first=True, dtype=int)
df_cleaned_encoded.columns
```

```python
df_cleaned_encoded = df_cleaned_encoded.rename(columns={
                    'DoHaiLong_Không Hài Lòng': 'DoHaiLong',
                    'LoaiKhachHang_KH Trung Thành': 'LoaiKhachHang',
                    'LoaiHinh_Đi Cùng Gia Đình/Bạn Bè': 'LoaiHinh'
                    })
df_cleaned_encoded.columns

```

```python
Y = df_cleaned_encoded['DoHaiLong']
X = df_cleaned_encoded.drop(columns=['DoHaiLong', 'TT'])
print(Y.shape, X.shape)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
tree = DecisionTreeClassifier(random_state=42)
tree.fit(X_train, Y_train)

tree_pred = tree.predict(X_test)
accuracy = accuracy_score(Y_test, tree_pred)
f1 = f1_score(Y_test, tree_pred)
recall = recall_score(Y_test, tree_pred)
precision = precision_score(Y_test, tree_pred)

print(f'Accuracy: {accuracy:.4f}')
print(f'F1 Score: {f1:.4f}')
print(f'Recall: {recall:.4f}')
print(f'Precision: {precision:.4f}')

from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))
plot_tree(tree, filled=True, feature_names=X.columns, class_names=['Hài Lòng', 'Không Hài Lòng'], max_depth=2)
plt.show()  
```

```python
X_test_10 = X_test.iloc[:10]
print("Predictions for first 10 samples in test set:")
predictions_10 = tree.predict(X_test_10)
print(predictions_10)
print("Actual labels for first 10 samples in test set:")
print(Y_test.iloc[:10].values)
```

```python
tree = DecisionTreeClassifier(random_state=42, max_depth=5, min_samples_split=10, min_samples_leaf=5)
tree.fit(X_train, Y_train)
tree_pred = tree.predict(X_test)
accuracy = accuracy_score(Y_test, tree_pred)
f1 = f1_score(Y_test, tree_pred)
recall = recall_score(Y_test, tree_pred)
precision = precision_score(Y_test, tree_pred)

print(f'Accuracy: {accuracy :.4f}')
print(f'F1 Score: {f1 :.4f}')
print(f'Recall: {recall :.4f}')
print(f'Precision: {precision :.4f}')
```

```python
from sklearn.model_selection import GridSearchCV
hyper_params = {
    'max_depth': [3, 5, 7, 10, 15, None],
    'min_samples_split': [2, 5, 10, 20, 45],
    'min_samples_leaf': [2, 5, 10, 20, 45]
    }

scoring = {'accuracy' : 'accuracy', 'f1' : 'f1', 'recall' : 'recall', 'precision' : 'precision'}
tree = DecisionTreeClassifier(random_state=42)

grid_search = GridSearchCV(estimator=tree, param_grid=hyper_params, scoring=scoring, refit='accuracy', cv=5)
grid_search.fit(X_train, Y_train)

```

```python
best_tree = grid_search.best_estimator_
tree_pred = best_tree.predict(X_test)
accuracy = accuracy_score(Y_test, tree_pred)
f1 = f1_score(Y_test, tree_pred)
recall = recall_score(Y_test, tree_pred)
precision = precision_score(Y_test, tree_pred)

print(f'Accuracy: {accuracy:.4f}')
print(f'F1 Score: {f1:.4f}')
print(f'Recall: {recall:.4f}')
print(f'Precision: {precision:.4f}')
```

```python
important_features = best_tree.feature_importances_
important_features = pd.Series(important_features, index=X.columns).sort_values(ascending=False)

plt.figure(figsize=(10,6))
important_features.head(10).plot(kind='bar')
plt.title('Top 10 Important Features')
plt.ylabel('Feature Importance Score')
plt.xlabel('Features')
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier(random_state=42, n_estimators=100)
forest.fit(X_train, Y_train)
forest_pred = forest.predict(X_test)
accuracy = accuracy_score(Y_test, forest_pred)
f1 = f1_score(Y_test, forest_pred)
recall = recall_score(Y_test, forest_pred)
precision = precision_score(Y_test, forest_pred)
print(f'Accuracy: {accuracy:.4f}')
print(f'F1 Score: {f1:.4f}')
print(f'Recall: {recall:.4f}')
print(f'Precision: {precision:.4f}')
```
