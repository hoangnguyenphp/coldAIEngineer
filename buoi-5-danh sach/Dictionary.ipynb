```python
# dictionary is collection of key-value pairs
# keys are unique an immutable
# value can be accessed via keys

en_vn = {"apple": "quả táo", "banana": "quả chuối"}

print (en_vn["apple"])

# check a key is in the list or not
print ("apple" in en_vn)

#insert new element
en_vn["mango"] = "quả xoài"
en_vn["durian"] = "quả xầu riêng"

print(en_vn)
# modify a value of an element
en_vn["durian"] = "quả sầu riêng"
print(en_vn)

# check  an element is in dictionary
print("mango" in en_vn)

# remove an element
fruit = en_vn.pop("banana")
print(en_vn, fruit)

```

```python
#get all keys
print(list(en_vn.keys()))

#get all values
print(list(en_vn.values()))

#get all items
print(list(en_vn.items()))

# loop an dictionary
for f in en_vn:
    print(f)

# loop all items
for item in list(en_vn.items()):
    en, vn = item
    print(en, vn)
```
