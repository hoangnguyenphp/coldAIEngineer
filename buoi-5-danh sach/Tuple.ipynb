```python
a = [1, 2,3, 4, 5] # List
b = (1 ,2, 3, 4, 5) # Tuple

# Tuple is smilar to List but we can not change the value, size, sort, remove, add  a tuple

#access by index
print(b[1])

for t in b:
    print(t)



```

```python
# 'tuple' object does not support item assignment
b[0] = 10
```

```python
# multiple
b = (1, 3, 5)
b = b*2
print(b)
```

```python
# pack
a = 1 ,2, 3
print (a)

# unpack
m, n, p = a
print(m)
print(n)
print(p)

```
