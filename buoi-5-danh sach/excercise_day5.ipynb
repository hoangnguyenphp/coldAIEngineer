4. Tìm số lớn nhất và nhỏ nhất trong danh sách:
Viết chương trình tìm và in ra số lớn nhất và số nhỏ nhất trong một danh sách các số nguyên.

```python
intList = [1 ,3 ,5, 6, 8, 8, 6]
min = intList[0]
max = intList[0]
for i in range(len(intList)):
    if(intList[i] < min):
        min = intList[i]
    if(intList[i] > max):
        max = intList[i]

print(f'Min number: {min} - Max number: {max}')
```

7. Ghép 2 danh sách:
Cho 2 danh sách toàn bộ là số nguyên với thứ tự đã được sắp xếp từ nhỏ dần đến lớn dần.
Hãy hợp nhất 2 danh sách này thành một danh sách duy nhất với thứ tự được sắp xếp từ nhỏ đến
lớn mà không dùng hàm sort cho List.

```python
sortedIntList1 = [2, 3, 3, 5, 8]
sortedIntList2 = [3,7, 9, 10]

sortedIntList1.extend(sortedIntList2)
print(sortedIntList1)

for i in range(len(sortedIntList1) - 1):
    indexOfMin = i
    for j in range(i, len(sortedIntList1)):
        if sortedIntList1[j] < sortedIntList1[indexOfMin]:
            indexOfMin = j
    temp = sortedIntList1[i]
    sortedIntList1[i] = sortedIntList1[indexOfMin]
    sortedIntList1[indexOfMin] = temp
print(sortedIntList1)
```

8. Lọc kết quả bầu cử:
Kết quả bầu cử là một danh sách các số nguyên dương, với mỗi số tương ứng với một ứng viên,
và mỗi phần tử của List tương ứng với một vote.
Hãy viết chương trình nhận vào list trên, sau đó tổng hợp số lượng phiếu bầu cho mỗi ứng viên
(Cho rằng mỗi ứng viên có ít nhất 1 phiếu bầu)
Ví dụ:
• Input: [1, 2, 3, 4, 4, 1, 1, 2, 2, 1]
• Output:
• Ứng viên thứ 1: 4 phiếu
• Ứng viên thứ 2: 3 phiếu
• Ứng viên thứ 3: 1 phiếu
• Ứng viên thứ 4: 2 phiếu

```python
voteList = [1, 2, 3, 4, 4, 1, 1, 2, 2, 1, 5, 9]
voteSet = set(voteList)

print(voteList)
print(voteSet)
result = {}

for i in voteSet:
    result[i] = voteList.count(i)

print(result)

for key, value in result.items():
    print(f'Candidate {key}: {value}')
```
