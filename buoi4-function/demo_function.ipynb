```python
def hello():
    print("Hello!")

hello()

def hi(name):
    print("Hello ", name) 

name = input("Enter a name:")
hi(name)

def add(a, b):
    c = a + b
    return c

print(add(10, 20))

def swap(a, b):
    c = a
    a = b
    b = c
    print(f"a = {a}, b = {b} inside swap")

a = 5
b =10
swap(a, b)
print(f"a = {a}, b = {b} outside swap")


def print_line(length):
    for i in range(length):
        print("*", end='')

def print_triagle(nline):
    for i in range(nline):
        print(print_line(i+1))

print_triagle(5)

def calculateN(n):
    result = 1
    for i in range(n):
        result = result * (i+1)
    return result

n = int(input("Enter a number for calculate n: "))

print(calculateN(5))

```
