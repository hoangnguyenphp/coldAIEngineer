1.Viết hàm tim_so_lon_nhat(a, b, c) nhận vào ba số và trả về số lớn nhất
trong ba số đó

```python
def findMax(a, b, c):
    max = a
    if (max < b):
        max = b
    if (max < c):
        max = c
    return max

a = int(input("Enter integer number a:"))
b = int(input("Enter integer number b:"))
c = int(input("Enter integer number c:"))

print(findMax(a, b, c))
```

Bài 2. Hàm chuyển đổi nhiệt độ:
Viết hàm chuyen_doi_nhiet_do(c) nhận vào nhiệt độ Celsius và chuyển
đổi sang Fahrenheit theo công thức:
F = (9/5) * C + 32

```python
def fahrenheitTemperatureConvert(celsiusTemperature):
    return (9/5) * celsiusTemperature + 32

celsiusTemperature = int(input("Input celsiusTemperature temperature:"))
print(fahrenheitTemperatureConvert(celsiusTemperature))
```

Bài 3. Viết hàm in hình chữ nhật.
Ví dụ nhập: n = 4, m = 5
in ra
* * * * *
* * * * *
* * * * *
* * * * *

```python
def printRectangle(length, width):
    for i in range(length):
        for j in range(width):
            print("*", end="")
        print()

length = int(input("Enter length:"))
width = int (input("Enter width:"))

printRectangle(length, width)
```

Bài 4. Viết hàm kiểm tra 1 số là số chính phương hay không

```python
def checkSquareNumber(intNumber):
    if intNumber == 1:
        return True
    for i in range(2, intNumber // 2 + 1):
        if i * i == intNumber:
            return True
        
    return False
        

intNumber = int(input("Enter an integer number:"))
print(checkSquareNumber(intNumber))
```

Bài 5. Viết hàm tính tiền điện
● Dưới 50 kWh: 1.678 VND/kWh
● Từ 51 đến 100 kWh: 1.734 VND/kWh
● Từ 101 đến 200 kWh: 2.014 VND/kWh
● Trên 200 kWh: 2.536 VND/kWh
Viết hàm main sử dụng vòng lặp yêu cầu người dùng nhập số điện tiêu
thụ, gọi hàm tính tiền rồi in ra giá phải trả, sau đó hỏi người dùng có
muốn tiếp tục không, nếu có thì quay lại từ đầu hỏi người dùng nhập
số điện.

```python
def caculatePaymentForElectricConsumation(electricConsummation):
    if electricConsummation <= 0:
        return 0.0
    
    consumationSteps = {1: 50, 2: 50, 3: 100}
    paymentSteps = {1: 1.678, 2: 1.734, 3: 2.014, 4: 2.536}
    totalPayment = 0.0
    step = 1

    while electricConsummation > 0:
        if step <= len(consumationSteps):
            if electricConsummation - consumationSteps[step] > 0:
                totalPayment += consumationSteps[step] * paymentSteps[step]
                electricConsummation = electricConsummation - consumationSteps[step]
                step = step + 1
            else: 
                totalPayment += electricConsummation * paymentSteps[step]
                electricConsummation = electricConsummation - consumationSteps[step]
        else:
            if electricConsummation > 0:
                totalPayment += electricConsummation * paymentSteps[step]
                electricConsummation = 0
    return totalPayment

cont = 'y'
while cont == 'y':
    electricConsummation = int(input("Enter the electric consummation:"))
    print(caculatePaymentForElectricConsumation(electricConsummation))
    cont = input("Do you want to continue (y/n)?")


```
