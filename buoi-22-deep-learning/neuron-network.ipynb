    02 Perceptron - Đơn vị cơ bản

        2.1. Perceptron: Đơn vị xử lý cơ bản

            Perceptron mô phòng một nơ-ron đơn lẻ với 3 bước xử lý:

                Tổng có trọng số
                Nhân đầu vào x với trọng số wy và cộng bias b

                    z = ∑ wixi + b

                    Trọng số (Weights - w): Mức độ quan trọng của đầu vào đối với quyết định cuối cùng.
                    Bias (b): Ngưỡng kích hoạt (dịch chuyển đường phân chia), giúp mô hình linh hoạt hơn.

            Hàm kích hoạt
                Ấp dụng hàm phi tuyến f lên : ŷ =f(z)

            Đầu ra
                Kết quả dự đoán ý.

        2.2. Biểu diễn Ma trận cho Perceptron
            Để tính toán hiệu quả trên máy tính, ta sử dụng đại số tuyến tính:
                z=wTx + b

                x: Vector đầu vào (cột)
                W: Vector trọng số (cột)
                b: Số vô hướng (Bias)

            Ví dụ tính toán

                Inputs: x = [2, 3, 1]T
                Weights: w = [0.5, -0.3, 0.8]T
                Bias: b = 0.1

                z = (0.5 × 2) + (-0.3 x 3) + (0.8 x 1) + 0.1
                z = 1.0 - 0.9 + 0.8 + 0.1

                => z= 1.0

                "Phép nhân ma trận cho phép tính toán song song hàng nghìn nơ-ron cùng lúc."

        2.3. Vai trò của Hàm kích hoạt (Activation Function)
            Chức năng Quyết định nơ-ron có "kích hoạt" hay không dựa trên tổng đầu vào z.

            Tại sao cần hàm kích hoạt: để tạo tính phi tuyến (non-linearity).
            
            Nếu không có, mạng nơ-ron dù nhiều lớp cũng chỉ tương đương một hàm tuyến tính đơn giản

            Ví dụ: Hàm bước (Step function)

                Nếu z ≥ 0 + Output = 1
                Nếu z < 0 + Output = 0

                Ứng dụng: Phân loại Spam (1) / Không (0)

    03 Kiến trúc mạng nơ-ron
        3.1. Kiến trúc Mạng Nơ-ron: 3 Loại tầng

            Input Layer (Tầng đầu vào): Nhận dữ liệu thô (ảnh, văn bản, số liệu). Số nơ-ron tương ứng với số đặc trưng (features) của dữ liệu.
            
            Hidden Layer(s) (Tầng ần) Nơi xày ra quá trình học và trích xuất đặc trưng. Mạng có nhiều tằng ần được gọi là Deep Learning. 

                Số lượng Perceptron ở tầng này thường là số mũ của 2: 2^1, 2^2, 2^3...
            
            Output Layer (Tầng đầu ra) Đưa ra kết quả dự đoán cuối cùng (xác suất, giá trị thực, hoặc phân loại).

            FULLY CONNECTED (DENSE) 
                Mọi nơ-ron ở tầng trước được kết nỗi với tất cả các nơ-ron ở tầng sau. Đây là dạng kết nối cơ bản nhất trong MLP.

        3.3. Chọn Hàm kích hoạt cho từng tầng

            Không phải tầng nào cũng dùng hàm giống nhau. Việc lựa chọn phụ thuộc vào vị trí và mục đích của tầng đó:


            TẮNG                            HÀM THƯỜNG DÙNG                      LÝ DO   

            Hidden Layers                    ReLU , Leaky ReLU                      Tính toán nhanh, tránh vanishing gradient, hội tụ 
            
                                                                                                tốt.
            
            Output (Hồi quy)                 Linear / Identity                      Cần giá trị liên tục bất kỳ.
            Output (Phân loại 2 lớp)         Sigmoid                                Đưa về xác suất (0, 1).
            Output (Phân loại nhiều lớp)     Softmax                                Phân phối xác suất tổng bằng 1.

    
    4. Feed-forward (Qua trình Tính xuôi)

        Định nghĩa: Dữ liệu đi một chiều từ Input -> Hidden -> Output. Không quay ngược lại.

                                Input + Hidden + Output

        Input x đưa vào tầng 1.

        Tỉnh tổng trọng số và áp dụng hàm kích hoạt.

        Kết quả trở thành input cho tầng tiếp theo.
        
        Lặp lại đến khi ra 9 ở Output layer.

        VÍ DỤ MINH HỌA
            Với mạng 2-2-1, đầu vào [1, 0.5] qua các phép nhân ma trận và ReLU/Sigmoid có thể cho ra dự đoán = 54.36%.

    5. Lan truyền ngược (Backpropagation): Ý tưởng

        5.1 Vấn đề Mạng dự đoán sai (có sự chênh lệch giữa ý và y). Làm sao để sửa?

        Giải pháp Tỉnh toán xem cần điều chỉnh mỗi trọng số ư bao nhiêu để giảm sai số.

        Cơ chế Tỉnh đạo hàm (gradient) của Loss theo từng trọng số, đi ngược từ Output -+ Input.

        HÀM LOSS (MẤT MÁT)
            Đo lường sai số dự đoán
            
            L= 1/2(ŷ-y)^2
            Ví dụ: Mean Squared Error (MSE)

        5.2. Gradient Descent & Chain Rule
        
            Gradient Descent (Tối ưu hóa): cập nhật trọng số để "đi xuống đồi" hàm Loss:

                w <- w - (α * ∂L/∂w)


            Chain Rule (Quy tắc chuỗi): để tính gradient ở các tằng sâu, ta nhân các đạo hàm thành phần
                ∂L/∂w = ∂L/∂ŷ * ∂ŷ/∂z * ∂z/∂w

            => Đây là trái tim giúp Deep Learning hoạt động.


    2x^2+5x−10

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.models import Sequential
import matplotlib.pyplot as plt


X = np.linspace(0, 10, 200)
Y = 3.0 * X + np.random.normal(0, 2, 200)

plt.scatter(X, Y)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

# định nghĩa model
model = Sequential(
    [
        Dense(1, input_shape=(1,), activation=None)
    ]
)

# cấu hình model, sgd = schokatic gradient discent, mse = mean square error
model.compile(optimizer="sgd", loss="mse")

# huấn luyện model
model.fit(X, Y, epochs=10, verbose=1)


loss_values = model.history.history['loss']
plt.plot(loss_values)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()

Y_pred = model.predict(X)
plt.scatter(X[::10], Y[::10], color='blue')
plt.scatter(X[::10], Y_pred[::10], color='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()


W, b = model.layers[0].get_weights()
print(W, b)
print(f'Y = {W[0][0]:.2f} * X + {b[0]:.2f}')


X = np.linspace(-20, 20, 200)
Y = 2.0 * X ** 2 + 5.0 * X - 10 + np.random.normal(-25, 25, 200)

plt.scatter(X, Y)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()


# định nghĩa model
model = Sequential(
    [
        Dense(1, input_shape=(1,), activation="tanh")
    ]
)
# cấu hình model
model.compile(optimizer="sgd", loss="mse")

# huấn luyện model
model.fit(X, Y, epochs=100, verbose=1)

# vẽ loss
loss_values = model.history.history['loss']
plt.plot(loss_values)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()



Y_pred = model.predict(X)
plt.scatter(X[::10], Y[::10], color='blue')
plt.scatter(X[::10], Y_pred[::10], color='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()


model = Sequential(
    [
        Dense(32, input_shape=(1,), activation="relu"),
        Dense(32, activation="relu"),
        Dense(32, activation="relu"),
        Dense(1, activation=None),
    ]
)
# cấu hình model
model.compile(optimizer="adam", loss="mse")

# huấn luyện model
model.fit(X, Y, epochs=200, verbose=1)

# vẽ loss
loss_values = model.history.history['loss']
plt.plot(loss_values)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()


Y_pred = model.predict(X)
plt.scatter(X[::10], Y[::10], color='blue')
plt.scatter(X[::10], Y_pred[::10], color='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()


Weights, b = model.layers[3].get_weights()
print(Weights, b)
print(len(Weights))
```
