```python
print("hello world")

a = 5
print(a, type(a))

a = 5.0
print(a, type(a))

a = "hello"
print(a, type(a))

# Array
a = [10, 20, 30, "hello", True]
print(a, type(a))

# Dictionary

vn_en_dictionary = {'hello': 'xin chào', 'cat': 'con mèo', 'dog': 'con chó'}
print(vn_en_dictionary['cat'])

# Get input from user
word = input('Enter English word:')
print(vn_en_dictionary[word])

# Type of input
name = input('Enter name: ')
num = input ('Enter a number: ')

print(name, type(name)) # Michael <class 'str'>
print(num, type(num)) # 10 <class 'str'> => input value is alway a string
```

```python

# Convert type
input_value = '10'
print(input_value, type(input_value))
# input_value = input_value + 3  => exception because input_value is a string
input_value = int(input_value) + 3
print(input_value, type(input_value))
```

```python
# print with separator
print ('value1', 'value2', 'value3', sep='-')

# output format
age = 10
name = 'Michael}'
value = 100
print (f'Name: {name}, Age: {age}, Value: {value}')

# print with additional space
value1 = 10
value2 = 100.635343
value3 = 'print something'
print(f'|{value1:<20}|')
print(f'|{value2:<20}|')
print(f'|{value3:<20}|')

# print with additional space and align
print(f'|{value1:<20}|')
print(f'|{value2:^20}|')
print(f'|{value3:>20}|')

# print decimal number with decimal places
print(f'{value2:.2f}')
print(f'{value2:20.2f}')
print(f'{value2:20.4f}')
```
