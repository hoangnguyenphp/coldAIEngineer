```python
# OPERATION FOR INTEGER

# plus minus
value1=10
value2=2
print ('Plus: ', value1 + value2)
print ('Minus: ', value1 - value2)
print ('Multiple: ', value1 * value2)
print ('Integer division: ', value1 // value2) #integer division
print ('Float division: ', value1 / value2) # float division
print ('Modules: ', value1 % value2) # modules
print ('Exponentiation: ', value1 ** value2) # exponentiation

# OPERATION FOR STRING
value1 = "Xin"
value2 = "Chào"

print (value1 + ' ' + value2)

# COMPARATIONS
print( 3 == 4)
print (3 != 4)
print(3 > 4)
print(3 < 4)
print(3 >= 4)
print(3 <=4)

```

```python
# LOGIC

#equal
print('=========equal')
print(3 == 3)
print(2 == 3)

#not equal
print('=========not equal')
print( 3 != 3)
print( 2 != 3)

#and
print('=========and')
print(True and True)
print(True and False)
print(False and False)

#or
print('=========or')
print(True or True)
print(True or False)
print(False or False)



#not (revert)
print('=========not (revert)')
print (not(3==3 and 4 ==4))
print (not(3==3 and 4 ==3))
```
